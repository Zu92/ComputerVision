```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
# cargar imagen RGB
imgRGB = cv.imread("CoordCrom_2.png")

# convertir la imagen a tipo flotante
imgRGB = imgRGB.astype(np.float32)

# sumar los valores de los canales R, G y B de cada pixel
sum_channels = np.sum(imgRGB, axis=2)

# dividir cada canal de cada pixel por la suma de los tres canales
img_whitepatch = imgRGB / sum_channels[:, :, np.newaxis]

# convertir la imagen a tipo entero de 8 bits
img_whitepatch = (img_whitepatch * 255).astype(np.uint8)

plt.imshow(img_whitepatch)
plt.show()
```

```python
# cargar imagen RGB
imgRGB = cv.imread("CoordCrom_1.png")

# convertir la imagen de RGB a espacio de color LAB
imgLAB = cv2.cvtColor(imgRGB, cv2.COLOR_BGR2RGB)

# separar los canales de color de la imagen LAB
L, A, B = cv2.split(imgLAB)

# ajustar los valores de los canales A y B en función de los valores máximos
L_max = np.max(L)
A_max = np.max(A)
B_max = np.max(B)

A = (A / A_max) * 255
B = (B / B_max) * 255
L = (L / L_max) *1

# combinar los canales de color ajustados para obtener la imagen procesada
img_whitepatch = cv2.merge([L, A, B])




plt.imshow(img_whitepatch)
plt.show()
```
