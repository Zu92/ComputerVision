```python

%matplotlib inline  

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Cargar una imagen en modo monocromático (un canal)
img = cv.imread('img1_tp.png', cv.IMREAD_GRAYSCALE)
img2 = cv.imread('img2_tp.png', cv.IMREAD_GRAYSCALE)
# Dimensión de la imagen
print(f'Tamaño imagen: {img.shape} pix')
print(img)
print(f'Tamaño imagen: {img2.shape} pix')
print(img)
```

```python
# Muestro la imagen con matplotlib
plt.figure()
plt.imshow(img)
plt.show()
plt.figure()
plt.imshow(img2)
plt.show()

# Qué pasó con la visualización ???
# Esto lo muestra en color porque matplot no sabe en que escala esta.
```

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=285)
plt.show()
plt.imshow(img2, cmap='gray', vmin=0, vmax=285)
plt.show()
```

```python
perfil = img2[:,150];
# Analizo la columna 150 justo el medio de la flor
# los petalos son blancos bajan intensidad, la parte del medio sube con el gris y vuelve a bajar con los petalos. Los bordes negros
plt.plot(perfil)
plt.show()
```

```python
# Ahora hagamos una marca en la imagen para indicar la línea que usamos para tomar el perfil
img_anotada = img2.copy()
img_anotada[:, 150]=0
plt.figure()
plt.imshow(img_anotada,cmap='gray')
plt.show()
```

```python

# Nueva figura
fig = plt.figure()

plt.subplot(211),plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.subplot(212),plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.show()

hist1, bins1 = np.histogram(img.ravel(), 256, [0, 256])



hist2,bins2 = np.histogram(img.ravel(),10,[0,256])



hist3,bins3 = np.histogram(img2.ravel(),256,[0,256])


hist4,bins4 = np.histogram(img2.ravel(),10,[0,256])



# Nueva figura
fig = plt.figure()

# Histogramas de la imagen
plt.subplot(221), plt.plot(hist1)
plt.subplot(222), plt.plot(hist2)
plt.subplot(223), plt.plot(hist3)
plt.subplot(224), plt.plot(hist4)
plt.show()

# Se puede observar que al ser imagenes en escala de grises, con valores en los extremos ambas imagenes se pueden reducir en un rangos de color corto
```
